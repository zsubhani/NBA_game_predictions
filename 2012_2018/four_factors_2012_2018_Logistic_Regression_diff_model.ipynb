```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import os
import numpy as np
```

```python
df = pd.read_csv('2012-18_teamBoxScore_diff_columns.csv')
df.head()
```

```python
# drop any rows with NaN values
df = df.dropna(how='any') 
```

```python
# only focus on the columns for the four factors
# put that into new dataframe
df1 = df[['diff_teamEFG%','diff_opptEFG%', 'diff_teamTO%', 'diff_opptTO%', 'diff_OREB%', 'diff_DREB%', 'diff_teamFTF', 'diff_opptFTF', 'outcome']]
df1.head()
```

```python
# Assign X (data) and y (target)
X = df1.drop("outcome", axis=1)
y = df1["outcome"]
print(X.shape, y.shape)
```

```python
#split our data into training and testing

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
# create a logistic regression model

from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
# Fit (train) or model using the training data

classifier.fit(X_train, y_train)
```

```python
# validate the model using the test data


print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
predictions = classifier.predict(X_test)
print(f"First 10 Predictions:   {predictions[:10]}")
print(f"First 10 Actual labels: {y_test[:10].tolist()}")
```

```python
# 0 stands for loss, 1 stands for win
pd.DataFrame({"Prediction": predictions, "Actual": y_test}).reset_index(drop=True)
```

```python
# create function to make predictions
def predict_outcome(road_team_abbr, road_team_stats, home_team_stats):
    road_team_array = np.array(road_team_stats)
    home_team_array = np.array(home_team_stats)
    diffs = road_team_array - home_team_array
    diffs_l = [diffs]
    prediction = classifier.predict(diffs_l)
    print(prediction)
    if ((prediction==1.).all()==True):
        print('Prediction is a Win for ' + str(road_team_abbr))
    else:
        print('Prediction is a Loss for ' + str(road_team_abbr))
```
